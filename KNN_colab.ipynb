```python
import numpy as np
import pandas as pd
```

```python
file_path = 'ConsumoCo2.csv'
data = pd.read_csv(file_path)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
x = data[['ENGINESIZE', 'CYLINDERS']]
y = data['CO2EMISSIONS']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"r2: {r2}")
print(f"(MSE): {mse}")
print(f"(MAE): {mae}")
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(x_test['ENGINESIZE'], y_test, color='blue', label='Dados Reais')
plt.plot(x_test['ENGINESIZE'], y_pred, color='red', linewidth=2, label='Reta de Regressão')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.title('Engine Size vs CO2 Emissions')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_test['CYLINDERS'], y_test, color='blue', label='Dados Reais')
plt.plot(x_test['CYLINDERS'], y_pred, color='red', linewidth=2, label='Reta de Regressão')
plt.xlabel('Cylinders')
plt.ylabel('CO2 Emissions')
plt.title('Cylinders vs CO2 Emissions')
plt.legend()

plt.tight_layout()
plt.show()
```
